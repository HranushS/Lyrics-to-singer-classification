Scraping lyricsfrom 'https://molorak.do.am/'


```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.common.keys import Keys
import time 
from bs4 import BeautifulSoup
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
```

```python
PATH = 'C:/Program Files (x86)/chromedriver.exe'
driver = webdriver.Chrome(PATH)
driver.get('https://molorak.do.am/')

artist_names =  ["Էմմի",'Ներսիկ Իսպիրյան',"Silva Hakobyan",'Անդրե',"Ruben Hakhverdyan",'Լիլիթ Հովհաննիսյան', 'Սիրուշո', 'Միհրան Ծառուկյան','Էրիկ', 'Նունե']

for artist_name in artist_names:
    wait = WebDriverWait(driver, 10)  
    search_input = wait.until(EC.presence_of_element_located((By.CSS_SELECTOR, 'form.hSearch input[name="q"]')))
    search_input.clear()
    search_input.send_keys(artist_name)
    search_button = driver.find_element(By.CSS_SELECTOR, 'form.hSearch input[type="submit"]')
    search_button.click()
    song_name_elements = driver.find_elements(By.CLASS_NAME, 'eTitle')

    num_songs = min(len(song_name_elements), 3)
    songs_found = 0
    i = 0

    while songs_found < num_songs and i < len(song_name_elements):
        song_link = song_name_elements[i].find_element(By.TAG_NAME, 'a')
        song_link.click()

        try:
            driver.implicitly_wait(5)
            lyrics_elements = driver.find_elements(By.CSS_SELECTOR, 'span[style="font-size:14px;"], span[style="font-size: 10pt;"]')
            if lyrics_elements:
                lyrics = '\n'.join(element.text for element in lyrics_elements)

                with open(f'lyrics_{artist_name}_{songs_found+1}.txt', 'w', encoding='utf-8') as file:
                    file.write(lyrics)

                print(f'Lyrics for {artist_name}, Song {songs_found+1}:')
                print(lyrics)

                songs_found += 1

        except NoSuchElementException:
            print(f'No lyrics found for {artist_name}, Song {i+1}')

        driver.back()
        song_name_elements = driver.find_elements(By.CLASS_NAME, 'eTitle')  
        i += 1

driver.quit()
```
